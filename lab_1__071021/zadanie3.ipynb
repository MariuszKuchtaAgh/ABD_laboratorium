Zdefiniowanie funkcji $f(x) = x^2 + 5$ i narysowanie jej wykresu dla przedziałów $(-1, 1)$, $(-6, 6)$, $(0, 5)$

```python
def f(x):
    return x**2 + 5

import numpy as np
import matplotlib.pyplot as plt

for plt_id, range_bounds in enumerate([(-1, 1), (-6, 6), (0, 5)]):
    range_i = np.linspace(range_bounds[0], range_bounds[1])
    vals_i = f(range_i)
    
    leg = "f(x)=x^2+5 dla zakresu (" + str(range_bounds[0]) + ', '+ str(range_bounds[1]) + ')'
    
    plt.figure(plt_id)
    plt.plot(range_i, vals_i, label = leg)
    plt.legend()
    plt.grid('on')
    plt.xlabel('Wartości x')
    plt.ylabel('Wartości f(x)')
```
